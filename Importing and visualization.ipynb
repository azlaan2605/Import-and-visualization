```python
import yfinance as yf
```

```python
data=yf.download('KO', start="2019-01-01", end="2021-03-31")
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline


plt.style.use('seaborn-darkgrid')
data['Close'].plot(figsize=(10, 7), color='green')
plt.title("Closing Price", fontsize=14)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(data['Close'], data['Open'])

# Set the title of the plot
plt.title('Open Vs Close', fontsize=14)
# Setting the x-axis title
plt.xlabel('BAC Price', fontsize=12)
# Setting the y-axis title
plt.ylabel('JPM Price', fontsize=12)
# Function for showing graph
plt.show()
```
